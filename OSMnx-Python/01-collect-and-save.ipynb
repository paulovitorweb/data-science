O [OSMnx](https://github.com/gboeing/osmnx) é um pacote para a linguagem de programação Python que permite fazer o download, modelar, analisar e visualizar a rede de ruas do OpenStreepMap. O objetivo do pacote, segundo o próprio criador da ferramenta, Geoff Boeing, é “tornar a coleta de dados e a criação e análise de redes de ruas simples, consistentes, automatizáveis e sonoras sob as perspectivas da teoria dos grafos, do transporte e do design urbano” (BOEING, 2017).

No conda, você pode instalar o pacote rodando o comando:

`conda install -c conda-forge osmnx`

Primeiro, vamos importar a biblioteca.


```python
import osmnx as ox
```

Para acessar a rede de ruas de um dado local, existe mais de uma forma. Você pode usar a função `graph_from_place` e fornecer como parâmetro uma string que represente um lugar. Na imagem abaixo, usei o método para o município de Campina Grande-PB e a função `plot_graph` para visualizar.

```python
ox.plot_graph(ox.graph_from_place('Campina Grande, Brazil'))
```

A seguinte, para o município de João Pessoa-PB.

```python
ox.plot_graph(ox.graph_from_place('João Pessoa, Brazil'))
```

É possível definir qual o tipo de via. No exemplo abaixo, acesso a rede de ruas públicas dirigíveis (`network_type=’drive’`) do bairro Mangabeira, em João Pessoa-PB.

```python
ox.plot_graph(ox.graph_from_place('Mangabeira, João Pessoa, Brazil', network_type='drive'))
```

Aqui, as ruas e os caminhos que os pedestres podem usar (`network_type=’walk’`) no bairro Manaíra, em João Pessoa-PB. Neste caso, ignora a direcionalidade unidirecional.

```python
ox.plot_graph(ox.graph_from_place('Manaíra, João Pessoa, Brazil', network_type='walk'))
```

Se você quiser salvar em shapefile para usar em algum SIG, pode fazê-lo com a função `save_graph_shapefile`. No caso abaixo, salvo a rede de ruas de João Pessoa e uso-as no QGIS. Na tabela de atributos você encontrará muitas informações úteis.

```python
ox.save_graph_shapefile(ox.graph_from_place('João Pessoa, Brazil'), filename='jp-streets')
```
